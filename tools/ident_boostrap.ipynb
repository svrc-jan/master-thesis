```python
import os
import sys

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
par_file = '/home/jsv/CVUT/master-thesis/data/ident_boostrap.csv'
df = pd.read_csv(par_file, header=None, names=['lambda_hor', 'lambda_ver', 'lambda_ang', 'offset_ang'])
```

```python
def print_stats(x):
	print("mean:   {:7.4f}\nmedian: {:7.4f}\nstd:    {:7.4f}\nCI95 l: {:7.4f}\nCI95 u: {:7.4f}\n".format(
	np.mean(x),
	np.median(x),
	np.std(x, ddof=1),
	np.quantile(x, 0.025),
	np.quantile(x, 1-0.025)
))
```

```python
x = df['offset_ang']
print_stats(x)
plt.hist(x)
plt.show()
```

```python
x = df['lambda_hor']
print_stats(x)
plt.hist(x)
plt.show()
```

```python
x = df['lambda_ver']
print_stats(x)
plt.hist(x)
plt.show()
```

```python
x = df['lambda_ang']
print_stats(x)
plt.hist(x)
plt.show()
```
